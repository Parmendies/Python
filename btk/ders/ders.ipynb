```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn import preprocessing


veriler = pd.read_csv('data.csv')
dummy_degiskenler = pd.get_dummies(veriler['ulke'],dtype='int')

dummy_degiskenler.columns = ['fr', 'tr', 'us']  

veriler = pd.concat([veriler, dummy_degiskenler], axis=1)
veriler.pop('ulke')
veriler['cinsiyet'] = (veriler['cinsiyet'] == 'e').astype(int)
veriler.dropna(inplace=True)

cinsiyet =  veriler.pop('boy')

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression



import statsmodels.api as sm
print(veriler)
print(veriler.values)
X = veriler.values 
X = np.append(arr=np.ones((X.shape[0], 1)).astype(int), values=X, axis=1)  
X = np.delete(X,2,axis=1)
linear = LinearRegression()
x_train, x_test, y_train, y_test = train_test_split(veriler, cinsiyet)
linear.fit(x_train,y_train)
y_pred = linear.predict(x_test)
sm_model = sm.OLS(cinsiyet,X).fit()
print(sm_model.summary())
```
